# 01 — Do extra local flat directions exist?

This notebook reads the same stored result files used by the README figures. It checks the first claim: the CNN has more locally output-preserving directions than the known architectural gauge predicts.

```python
from pathlib import Path
import json
import numpy as np
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    ROOT = ROOT.parent
RESULTS = ROOT / 'results'
print('repo root:', ROOT)
```

```python
with open(RESULTS / 'walk_test_cnn.json') as f:
    records = json.load(f)
rec = records[0]
print('anchor:', rec['ckpt'])
print('known architectural gauge:', rec['gauge_dim'])
print('measured output-preserving dimension:', rec['tangent_dim'])
print('measured extras:', rec['extra_dim'])
```

```python
npz_path = RESULTS / rec.get('npz', 'tangent_cnn_CNN_000.npz')
if not npz_path.exists():
    npz_path = RESULTS / 'tangent_cnn_CNN_000.npz'
z = np.load(npz_path)
sigmas = z['S']
gauge_dim = rec['gauge_dim']
extra_dim = rec['extra_dim']
P = len(sigmas)

fig, ax = plt.subplots(figsize=(8, 4))
ax.semilogy(np.arange(1, P + 1), np.clip(sigmas, 1e-17, None), lw=1.2)
ax.axvline(P - gauge_dim, color='black', ls='--', lw=1, label=f'gauge cutoff P-{gauge_dim}')
ax.axvspan(P - gauge_dim, P, color='#5B8E7D', alpha=0.10, label='known gauge')
ax.axvspan(P - gauge_dim - extra_dim, P - gauge_dim, color='#D97757', alpha=0.10, label='measured extras')
ax.set_xlabel('singular-value index')
ax.set_ylabel('singular value of J')
ax.set_title('CNN Jacobian spectrum: extra near-zero directions')
ax.grid(alpha=0.3)
ax.legend()
plt.tight_layout()
```

Interpretation: the green band is the 24-dimensional channel-rescaling gauge that the architecture already explains. The orange band is the additional locally flat subspace measured around this trained CNN anchor. This is not yet a hidden-symmetry claim; it is the measurable phenomenon this project studies.
