# 03 — Are the extra directions just a finite-sample artifact?

If the Jacobian is measured on too few data points, it can have artificial null directions. This notebook reads the tangent-dimension scaling sweep and checks whether the measured flat dimension collapses to the 24-dimensional architectural gauge as more samples are used.

```python
from pathlib import Path
import json
import numpy as np
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    ROOT = ROOT.parent
RESULTS = ROOT / 'results'
print('repo root:', ROOT)
```

```python
with open(RESULTS / 'tangent_dim_scaling.json') as f:
    scaling = json.load(f)
print('anchors:', [a['anchor'] for a in scaling])
```

```python
fig, ax = plt.subplots(figsize=(8, 4.5))
markers = ['o', 's', '^', 'D', 'v']
for i, anchor in enumerate(scaling):
    rows = [r for r in anchor['records'] if 'tangent_dim' in r]
    ns = np.array([r['n_samples'] for r in rows])
    dims = np.array([r['tangent_dim'] for r in rows])
    ax.plot(ns, dims, '-' + markers[i % len(markers)], lw=2, ms=6, label=anchor['anchor'])

gauge_dim = scaling[0]['records'][0]['gauge_dim']
ax.axhline(gauge_dim, color='#5B8E7D', ls='--', lw=2, label=f'known gauge = {gauge_dim}')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Jacobian sample count')
ax.set_ylabel('measured output-preserving dimension')
ax.set_title('Flat dimension shrinks with more data, then remains above gauge')
ax.grid(alpha=0.3, which='both')
ax.legend(ncol=2)
plt.tight_layout()
```

Interpretation: the count drops sharply when the sample size increases, which confirms that some early directions were sample artifacts. The important point is that the dimension does not collapse to the 24-dimensional architectural gauge in these stored runs.
