# 04 — Why the first Lie-bracket attempt is inconclusive

This notebook records the negative methodological result. The question was whether the measured flat directions behave like directions generated by a hidden smooth symmetry. The first projected-vector-field test fails even on the MLP gauge sanity case, so it should not be used as evidence against a hidden structure in the CNN.

```python
from pathlib import Path
import json
import numpy as np
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if not (ROOT / 'src').exists():
    ROOT = ROOT.parent
RESULTS = ROOT / 'results'
print('repo root:', ROOT)
```

```python
with open(RESULTS / 'lie_test_mlp.json') as f:
    lie = json.load(f)
print('anchors:', [a['anchor'] for a in lie])
print('first spectral diagnostic:', lie[0].get('spectral_gap', lie[0].get('spectral_diag')))
```

```python
closed = []
direct = []
for anchor in lie:
    closed.extend(anchor['records'].get('closed_loop', []))
    direct.extend(anchor['records'].get('direct_bracket', []))

fig, axes = plt.subplots(1, 2, figsize=(9, 4), sharey=True)
for ax, rows, key, xlabel, title in [
    (axes[0], closed, 'eps', 'epsilon', 'closed-loop test'),
    (axes[1], direct, 'h', 'finite-difference h', 'direct bracket test'),
]:
    groups = {}
    for r in rows:
        groups.setdefault(r[key], []).append(r['in_frac'])
    xs = sorted(groups)
    ax.boxplot([groups[x] for x in xs], positions=np.arange(len(xs)), widths=0.55, patch_artist=True)
    ax.axhline(1.0, color='#5B8E7D', ls='--', lw=1, label='would mean inside')
    ax.axhline(0.0, color='#D97757', ls=':', lw=1, label='mostly outside')
    ax.set_xticks(np.arange(len(xs)))
    ax.set_xticklabels([f'{x:.0e}' for x in xs])
    ax.set_xlabel(xlabel)
    ax.set_title(title)
    ax.grid(alpha=0.3)
axes[0].set_ylabel('fraction inside measured flat subspace')
axes[0].legend()
plt.tight_layout()
```

Interpretation: this test uses projected vector fields, not true vector fields that follow the unknown equivalence sheet. The low inside-fraction therefore reflects a defect of the test construction. This should be treated as a useful negative result: the next project should construct better tangent-following fields or move the question into representation space.
